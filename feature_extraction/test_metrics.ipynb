```python
import torch
import numpy as np
import os
```

```python
real_folder = "/scratch/toskov/data/geom_align_outputs/"
real_files = os.listdir(real_folder)

gen_folder = "/scratch/toskov/data/geom_align_outputs_gen/"
gen_files = os.listdir(gen_folder)
```

```python
real_means = []
n_missing = 0
n_good = 0
real_mean_list = []
real_max_list = []
real_median_list = []
for real in real_files:
    folder = os.path.join(real_folder, real)
    if not os.path.isdir(folder):
        continue
    
    metrics_file = os.path.join(folder, "metrics.pth")
    if not os.path.exists(metrics_file):
        n_missing += 1
        continue
    
    metrics = torch.load(metrics_file)
    n_good += 1
    
    # Has n_frames elements, each of which has (up to) window_size elements.
    # Average over the windows and then over the frames
    
    #do this for a bunch of metrics
    
    mean = np.mean([np.mean([x["mean"] for x in frame]) for frame in metrics])
    real_mean_list.append(mean)
    
    # max_value = np.max([np.max([x["mean"] for x in frame]) for frame in metrics])
    # real_max_list.append(max_value)
    
    # median = np.median([np.median([x["mean"] for x in frame]) for frame in metrics])
    # real_median_list.append(median)
    


print(f"Number of missing files: {n_missing}")
print(f"Number of good files: {n_good}")
print(real_mean_list)
```

```python
gen_means = []
n_missing = 0
n_good = 0
gen_mean_list = []
gen_max_list = []
gen_median_list = []
for gen in gen_files:
    folder = os.path.join(gen_folder, gen)
    if not os.path.isdir(folder):
        continue
    
    metrics_file = os.path.join(folder, "metrics.pth")
    if not os.path.exists(metrics_file):
        n_missing += 1
        continue
    
    metrics = torch.load(metrics_file)
    n_good += 1
    
    # Has n_frames elements, each of which has (up to) window_size elements.
    # Average over the windows and then over the frames
    
    #do this for a bunch of metrics
    
    mean = np.mean([np.mean([x["mean"] for x in frame]) for frame in metrics])
    gen_mean_list.append(mean)
    
    # max_value = np.max([np.max([x["mean"] for x in frame]) for frame in metrics])
    # gen_max_list.append(max_value)
    
    # median = np.median([np.median([x["mean"] for x in frame]) for frame in metrics])
    # gen_median_list.append(median)
    


print(f"Number of missing files: {n_missing}")
print(f"Number of good files: {n_good}")
print(gen_mean_list)
```

```python
import matplotlib.pyplot as plt
from sklearn.metrics import roc_curve, auc

# start with means

# Combine data
X = np.array(real_mean_list + gen_mean_list)
y = np.concatenate([np.zeros_like(real_mean_list), np.ones_like(gen_mean_list)])

mask = np.isnan(X)

X = X[~mask]
y = y[~mask]
# Ensure that the lengths of X and y match
if len(X) != len(y):
    raise ValueError("Length of X and y do not match after filtering NaNs.")

# Compute ROC curve and ROC area
fpr, tpr, thresholds = roc_curve(y, X)  # Use values as scores
roc_auc = auc(fpr, tpr)

# Plot ROC curve
plt.rcParams.update({'font.size': 20})
plt.figure(figsize=(8, 6))
plt.plot(fpr, tpr, color='darkorange', lw=2, label=f'ROC curve (AUC = {roc_auc:.2f})')
plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--', label='Chance level')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic')
plt.legend(loc="lower right")
plt.grid(True)
# plt.show()
# make tight
plt.tight_layout()
# increase font size

plt.savefig("roc_curve_means.pdf")
plt.show()
```

```python
from sklearn.metrics import precision_recall_curve, average_precision_score

# Compute PR curve
precision, recall, thresholds = precision_recall_curve(y, X)
ap_score = average_precision_score(y, X)

# Compute F1 scores (excluding the last point, which is for threshold=None)
f1_scores = 2 * precision[:-1] * recall[:-1] / (precision[:-1] + recall[:-1] + 1e-8)
best_idx = np.argmax(f1_scores)
best_threshold = thresholds[best_idx]
best_precision = precision[best_idx]
best_recall = recall[best_idx]
best_f1 = f1_scores[best_idx]

# Plot PR curve
plt.figure(figsize=(8, 6))
plt.plot(recall, precision, color='blue', lw=2, label=f'PR curve (AP = {ap_score:.2f})')
plt.scatter(best_recall, best_precision, color='red', label=f'Best F1 = {best_f1:.2f}\nThreshold = {best_threshold:.2f}', zorder=5)
plt.xlabel('Recall')
plt.ylabel('Precision')
plt.title('Precision-Recall Curve')
plt.legend(loc="lower center")
plt.grid(True)
# plt.show()
plt.tight_layout()
plt.savefig("pr_curve.pdf")
```
